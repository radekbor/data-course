```python
import pandas as pd
```

Importowanie danych z tdf

```python
## stworz df z 
url = "tour_de_france_1.csv"

df = pd.read_csv(url)
df
```

```python
# Narysuje wykres dla kolumny "Winner's avg speed"
avg_speed_column_name = "Winner's avg speed"
df[avg_speed_column_name].plot()
```

```python
# Narysuje wykres dla kolumny "Total distance (km)"
todal_distance_column_name = "Total distance (km)"
df[todal_distance_column_name].plot()
```

```python
import locale
locale.setlocale( locale.LC_ALL, 'en_US.UTF-8' ) 

distance_converter = lambda x: locale.atof(x.replace('"', '').replace("\t", '').strip())


#importuj jeszcze raz przy pomocy converera
converters = {todal_distance_column_name: distance_converter}
df = pd.read_csv(url, converters = converters)
df
```

## Wykresy predkosci oraz dystansu

```python
import matplotlib.pyplot as plt
df[avg_speed_column_name].plot(label = "avg speed")
plt.legend()
plt.show()
df[todal_distance_column_name].plot(label = "distance")
plt.legend()
```

## Wykres procent zawodnikow ktory ukonczyli

```python
(df["Finishers"]/df["Entrants"] * 100).plot()
import numpy as np
last = (df["Year"][len(df)-1])
labels = np.arange(df["Year"][0], last, step = 10)
ticks = labels - df["Year"][0]
plt.xticks(ticks = ticks, labels = labels)
```

```python
labels = []
values = []
for (name, group) in df.groupby("Starting country"):
    labels.append(name)
    values.append(len(group))

dataframe = pd.DataFrame(values, index = labels)
dataframe.plot.bar()
```
