## Histogram

```python
import numpy as np
```

```python
np.random.randn(1000)
```

```python
iq = 100 + 15 * np.random.randn(10000)
```

```python
plt.hist(iq, 20, density=1, facecolor='g', alpha=0.95)
plt.title('IQ Histogram')
```

### Bar + stack

```python
labels = ['A', 'B']

g1 = [10, 20]
g2 = [10, 40]
g3 = [10, 60]

fig, ax = plt.subplots()

ax.bar(labels, g1, 0.3, label='g1')
ax.bar(labels, g2, 0.3, bottom=g1,
       label='g2')
ax.bar(labels, g3, 0.3, bottom=g2,
       label='g3')
```

```python
labels = ['A', 'B']

g1 = np.array([10, 20])
g2 = np.array([10, 40])
g3 = np.array([10, 60])

fig, ax = plt.subplots()

ax.bar(labels, g1, 1, label='g1')
ax.bar(labels, g2, 0.9, bottom=g1,
       label='g2')
ax.bar(labels, g3, 0.8, bottom=g1 + g2,
       label='g3')
```

## IRIS 

```python
iris = np.genfromtxt("iris.csv",  delimiter=',')
iris
```

```python
iris[:,4]
```

```python
colours = ['red', 'orange', 'blue']
species = ['I. setosa', 'I. versicolor', 'I. virginica']
for i in range(0, 3):    
    species_df = iris[iris[:,4] == i] 
    ratio = species_df[:,1] / species_df[:,0]
    plt.scatter(        
        species_df[:,0],        
        species_df[:,1],
        color=colours[i],        
        alpha=0.5,        
        label=species[i]   
    )
```
